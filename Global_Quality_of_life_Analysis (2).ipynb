## Data Filtering

```python
# Top 10 Countries by Quality of Life Score
df['Quality of Life Value'] = pd.to_numeric(df['Quality of Life Value'], errors='coerce')

top_qol = df.sort_values(by='Quality of Life Value', ascending=False).head(10)
print(top_qol[['country', 'Quality of Life Value']])
```

```python
#Average Score for Each Metric
df[['Purchasing Power Value', 'Safety Value', 'Health Care Value',
    'Climate Value', 'Cost of Living Value', 'Traffic Commute Time Value',
    'Pollution Value', 'Quality of Life Value']].mean()
```

```python
#Find Countries with Very High Safety
df[df['Safety Category'] == "'Very High'"][['country', 'Safety Value']]
```

```python
# Best Countries with High Safety and Low Pollution
best_countries = df[(df['Safety Value'] > 70) & (df['Pollution Value'] < 30)]
print(best_countries[['country', 'Safety Value', 'Pollution Value']])
```

```python
df.describe(include='all')
```

```python
# Remove colon (:) and convert to float
df['Quality of Life Value'] = df['Quality of Life Value'].replace(r":\s*", "", regex=True).astype(float)
```

```python
df.replace(0, np.nan, inplace=True)
```

```python
#Drop rows with >50% missing values
df = df[df.isnull().mean(axis=1) < 0.5]
```

```python
#Encode Ordinal Categories
category_map = {
    'Very Low': 1,
    'Low': 2,
    'Moderate': 3,
    'High': 4,
    'Very High': 5,
    0: np.nan  # Treat 0 as missing
}

category_cols = [col for col in df.columns if 'Category' in col]

for col in category_cols:
    df[col] = df[col].replace(category_map)
```

```python
# Filter countries where pollution is very high and safety is low
high_risk_countries = df[(df['Pollution Value'] > 75) & (df['Safety Value'] < 40)]
high_risk_countries
```

```python
# Filter countries where people earn less but living costs are high
costly_countries = df[(df['Purchasing Power Value'] < 50) & (df['Cost of Living Value'] > 50)]

costly_countries
```

```python
# Filter for countries offering good quality of life and low commute burden
ideal_living = df[(df['Quality of Life Value'] > 80) & (df['Traffic Commute Time Value'] < 25)]

ideal_living
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Clean 'Quality of Life Value' using regex extraction
def extract_numeric(value):
    match = re.search(r"(\d+\.?\d*)", str(value))
    if match:
        return float(match.group(1))
    return None

df['Quality of Life Value'] = df['Quality of Life Value'].apply(extract_numeric)

# Columns to visualize
numeric_columns = [
    'Purchasing Power Value',
    'Safety Value',
    'Health Care Value',
    'Climate Value',
    'Cost of Living Value',
    'Property Price to Income Value',
    'Traffic Commute Time Value',
    'Pollution Value',
    'Quality of Life Value'
]

# Ensure all columns are numeric
for col in numeric_columns:
    if df[col].dtype == object:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Prepare side-by-side barplots in a compact grid
fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.flatten()

for idx, col in enumerate(numeric_columns):
    temp_df = df[['country', col]].dropna()
    temp_df = temp_df[temp_df[col] > 0]
    top_countries = temp_df.sort_values(by=col, ascending=False).head(10)

    sns.barplot(
        x=col,
        y='country',
        data=top_countries,
        ax=axes[idx],
        palette="magma"
    )
    axes[idx].set_title(f'Top 10 Countries by {col}', fontsize=11)
    axes[idx].set_xlabel(col, fontsize=9)
    axes[idx].set_ylabel('Country', fontsize=9)

plt.show()
```

```python
#To compare trends of different quality metrics across countries.

def extract_numeric(value):
    match = re.search(r"(\d+\.?\d*)", str(value))
    if match:
        return float(match.group(1))
    return None
df['Quality of Life Value'] = df['Quality of Life Value'].apply(extract_numeric)

# Columns to compare
cols = ['Purchasing Power Value', 'Safety Value', 'Health Care Value', 'Quality of Life Value']
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df_clean = df[['country'] + cols].dropna().head(10)

# Line Plot
plt.figure(figsize=(10,6))
for col in cols:
    plt.plot(df_clean['country'], df_clean[col], marker='o', label=col)
plt.xticks(rotation=90)
plt.title("Line Plot: Comparison of Metrics Across Countries")
plt.xlabel("Country")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
#To examine the relationship between Pollution and Quality of Life
df['Quality of Life Value'] = df['Quality of Life Value'].apply(extract_numeric)

cols = ['Pollution Value', 'Quality of Life Value']
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df_clean = df[['country'] + cols].dropna().head(20)

plt.figure(figsize=(7,5))
plt.scatter(df_clean['Pollution Value'], df_clean['Quality of Life Value'], color='teal')
plt.xlabel("Pollution Value")
plt.ylabel("Quality of Life Value")
plt.title("Scatter Plot: Pollution vs Quality of Life")
plt.tight_layout()
plt.show()
```

```python
#Visualize Safety Values clearly across countries.
df['Safety Value'] = pd.to_numeric(df['Safety Value'], errors='coerce')

df_clean = df[['country', 'Safety Value']].dropna().tail(10)

# Stem Plot
plt.stem(df_clean['country'], df_clean['Safety Value'])
plt.xticks(rotation=45)
plt.title("Stem Plot: Safety Values by Country")
plt.xlabel("Country")
plt.ylabel("Safety Value")
plt.show()
```

```python
#See the distribution of Purchasing Power across countries.
df['Purchasing Power Value'] = pd.to_numeric(df['Purchasing Power Value'], errors='coerce')

df_clean = df['Purchasing Power Value'].dropna()

# Histogram
plt.hist(df_clean, bins=10, color='orange', edgecolor='black')
plt.title("Histogram: Purchasing Power Distribution")
plt.xlabel("Purchasing Power Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

```

```python

# Clean 'Quality of Life Value' robustly using regex to extract number

df['Quality of Life Value'] = df['Quality of Life Value'].apply(extract_numeric)

# Drop rows where 'Quality of Life Value' or 'country' is missing
df_clean = df[['country', 'Quality of Life Value']].dropna()

# Check if we have at least 5 countries for the pie chart
if len(df_clean) < 5:
    print("Not enough data for a pie chart. Please check your CSV data.")
else:
    # Select top 5 countries by Quality of Life Value
    top5 = df_clean.sort_values(by='Quality of Life Value', ascending=False).head(5)

    # Plot Pie Chart
    plt.figure(figsize=(7,7))
    plt.pie(
        top5['Quality of Life Value'],
        labels=top5['country'],
        autopct='%1.1f%%',
        startangle=140
    )
    plt.title("Pie Chart: Quality of Life Share (Top 5 Countries)")
    plt.tight_layout()
    plt.show()
```

```python
#Check the spread and outliers in Health Care Values.

df['Health Care Value'] = pd.to_numeric(df['Health Care Value'], errors='coerce')

df_clean = df['Health Care Value'].dropna()

# Box Plot
plt.boxplot(df_clean)
plt.title("Box Plot: Health Care Value Distribution")
plt.ylabel("Health Care Value")
plt.tight_layout()
plt.show()

```

```python
#top countries for highest quality of life
# Clean 'Quality of Life Value'
df['Quality of Life Value'] = df['Quality of Life Value'].apply(extract_numeric)

# Drop missing and zero/negative values
df_clean = df.dropna(subset=['Quality of Life Value'])
df_clean = df_clean[df_clean['Quality of Life Value'] > 0]

# Select top 20 countries with highest Quality of Life
top_qol_df = df_clean[['country', 'Quality of Life Value']].sort_values(
    by='Quality of Life Value', ascending=False).head(20)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(
    y='country',
    x='Quality of Life Value',
    data=top_qol_df,
    palette='viridis'
)

plt.title('Top 20 Countries with Highest Quality of Life')
plt.xlabel('Quality of Life Value')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
